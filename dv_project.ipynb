## DownSampling

```python
# Assuming 'df' is your DataFrame and 'target' is your target column
majority_class = 1
minority_class = 0

new_df = df[df["labels"]==2]
# Separate majority and minority classes
df_majority = df[df.labels==majority_class]
df_minority = df[df.labels==minority_class]

# Downsample majority class
from sklearn.utils import resample
df_majority_downsampled = resample(df_majority,
                                 replace=False,    # sample without replacement
                                 n_samples=len(df_minority),     # to match minority class
                                 random_state=123) # reproducible results

# Combine minority class with downsampled majority class
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_downsampled = pd.concat([df_downsampled,new_df],axis=0)
# Display new class counts
print(df_downsampled.labels.value_counts())
```

```python
df_downsampled['labels'].value_counts().plot.bar()
plt.show()
```

```python
x = df_downsampled.drop(["labels"],axis=1)
y = df_downsampled["labels"]
```

```python
plt.boxplot(df_downsampled)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=99, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier


DT_entropy = DecisionTreeClassifier(random_state=50)
DT_entropy.fit(X_train,y_train)
```

```python
y_pred = DT_entropy.predict(X_test)


# Compute classification metrics
matrix(DT_entropy,y_pred)
```

## OverSampling

```python
from sklearn.utils import resample

# Assuming 'df' is your DataFrame and 'target' is your target column
majority_class = 1
minority_class = 0
new_df1 = df[df["labels"]==2]
# Separate majority and minority classes
df_majority = df[df.labels==majority_class]
df_minority = df[df.labels==minority_class]

# Upsample minority class
df_minority_upsampled = resample(df_minority,
                                 replace=True,     # sample with replacement
                                 n_samples=len(df_majority),    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled,new_df1],axis=0)


majority_class = 1
minority_class = 2
new_df1 = df_upsampled[df_upsampled["labels"]==0]
df_upsampled = df_upsampled[df_upsampled["labels"]!=0]
# Separate majority and minority classes
df_majority = df_upsampled[df_upsampled.labels==majority_class]
df_minority = df_upsampled[df_upsampled.labels==minority_class]

# Upsample minority class
df_minority_upsampled = resample(df_minority,
                                 replace=True,     # sample with replacement
                                 n_samples=len(df_majority),    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled,new_df1],axis=0)

# Display new class counts
print(df_upsampled.labels.value_counts())
```

```python
df_upsampled['labels'].value_counts().plot.bar()
plt.show()
```

```python
x = df_upsampled.drop(["labels"],axis=1)
y = df_upsampled["labels"]
```

```python
plt.boxplot(df_upsampled)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1, stratify=y)
```

### Logestic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='lbfgs',random_state=99)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
matrix(lr,y_pred)
```
